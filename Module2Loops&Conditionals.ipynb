4.1 Using if/elif/else statements to guess a number

```python
secret = 5
guess = 7

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

4.2 Utilizing two variables and if/elif/else statements to identify a item. 

```python
small = True
green = True

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

6.1 Using a for loop to print the number of a list

```python
list = [3, 2, 1, 0]
for number in list:
    print(number)
```

```python
6.2 Assigns 7 to guess_me, then utilizing a while loop
```

```python
guess_me = 7
number = 1

while number <= guess_me:
    if number < guess_me:
        print('too low')
    else:
        print('found it!')
        break
    number += 1
else:
    print('oops')

        
```

6.3 Assigning the number 5 to guess_me, then utilizing a for loop

```python
guess_me = 5
for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
else:
    print('oops')

    
```
